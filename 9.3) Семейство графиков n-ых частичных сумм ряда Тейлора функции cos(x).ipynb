# 9.3) Семейство графиков n-ых частичных сумм ряда Тейлора функции cos(x)

$$\cos(x)=1-\frac{x^2}{2!}+\frac{x^4}{4!}-\frac{x^6}{6!}+...$$

```julia
using Plots

FloatN  = Union{Float16,Float32,Float64}
NumberN = Union{FloatN, Complex{FloatN}}

function sum_cos(x, n)
    sn = typeof(x)(0)
    an = 1
    m = 2
    for i in 1:n
        sn += an
        an *= (-1) * x^2 / m / (m - 1)
        m += 2
    end
    return sn
end

x = 0:0.01:2*pi
y = Float64[]
for i in x
    push!(y, sum_cos(i, 2))
end
plot(x, y, title = "Семейство графиков n-ых частичных сумм", label = "n = 2", legend = :outertopleft)
# n - кол-во слагаемых

y1 = Float64[]
for i in x
    push!(y1, sum_cos(i, 4))
end
plot!(x, y1, label = "n = 4")

y2 = Float64[]
for i in x
    push!(y2, sum_cos(i, 8))
end
plot!(x, y2, label = "n = 8")

y3 = Float64[]
for i in x
    push!(y3, sum_cos(i, 16))
end
plot!(x, y3, label = "n = 16")
```
